```python
def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def is_winner(board, player):
    for i in range(3):
        if all(board[i][j] == player for j in range(3)) or all(board[j][i] == player for j in range(3)):
            return True

    if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
        return True

    return False

def is_board_full(board):
    return all(all(cell != ' ' for cell in row) for row in board)

def evaluate(board):
    if is_winner(board, 'X'):
        return 1
    elif is_winner(board, 'O'):
        return 'O'
    elif is_board_full(board):
        return 0
    else:
        return None

def minimax(board, depth, maximizing_player):
    scores = {'X': 1, 'O': -1, 'tie': 0}

    result = evaluate(board)
    if result is not None:
        return scores[result]

    empty_cells = [(i, j) for i in range(3) for j in range(3) if board[i][j] == ' ']

    if maximizing_player:
        max_eval = float('-inf')
        for i, j in empty_cells:
            board[i][j] = 'X'
            eval_score = minimax(board, depth + 1, False)
            board[i][j] = ' '  # Undo the move
            max_eval = max(max_eval, eval_score)
        return max_eval
    else:
        min_eval = float('inf')
        for i, j in empty_cells:
            board[i][j] = 'O'
            eval_score = minimax(board, depth + 1, True)
            board[i][j] = ' '  # Undo the move
            min_eval = min(min_eval, eval_score)
        return min_eval

def get_best_move(board):
    best_val = float('-inf')
    best_move = None

    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                board[i][j] = 'X'
                move_val = minimax(board, 0, False)
                board[i][j] = ' '  # Undo the move

                if move_val > best_val:
                    best_val = move_val
                    best_move = (i, j)

    return best_move

def main():
    board = [[' ' for _ in range(3)] for _ in range(3)]
    current_player = 'X'

    while True:
        print_board(board)

        if current_player == 'X':
            print("Enter your move (row and column separated by space): ")
            row, col = map(int, input().split())
            if 0 <= row < 3 and 0 <= col < 3 and board[row][col] == ' ':
                board[row][col] = 'X'
            else:
                print("Invalid move. Try again.")
                continue
        else:
            print("Calculating computer's move...")
            row, col = get_best_move(board)
            board[row][col] = 'O'

        if is_winner(board, current_player):
            print_board(board)
            print(f"{current_player} wins!")
            break
        elif is_board_full(board):
            print_board(board)
            print("It's a tie!")
            break

        current_player = 'O' if current_player == 'X' else 'X'

if __name__ == "__main__":
    main()
```

```python
# Tic Tac Toe using MiniMax Algorithm

# Function to print the board
def print_board(board):
    print("-------------")
    for i in range(3):
        print("| " + " | ".join(board[i]) + " |")
        print("-------------")

# Function to check if the game is over
def game_over(board):
    # Check rows
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != ' ':
            return True
    
    # Check columns
    for i in range(3):
        if board[0][i] == board[1][i] == board[2][i] != ' ':
            return True
    
    # Check diagonal
    if board[0][0] == board[1][1] == board[2][2] != ' ' or board[0][2] == board[1][1] == board[2][0] != ' ':
        return True
    
    # Check if the board is full
    for row in board:
        if ' ' in row:
            return False
    
    return True

# Function to calculate the score of the board
def calculate_score(board):
    if game_over(board):
        if check_winner(board, 'X'):
            return -1
        elif check_winner(board, 'O'):
            return 1
        else:
            return 0

# Minimax algorithm function
def minimax(board, depth, is_maximizing):
    score = calculate_score(board)
    
    if score == 1:
        return 1
    if score == -1:
        return -1
    if score == 0:
        return 0
    
    if is_maximizing:
        best_score = float('-inf')
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'O'
                    score = minimax(board, depth+1, False)
                    board[i][j] = ' '
                    best_score = max(score, best_score)
        return best_score
    
    else:
        best_score = float('inf')
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'X'
                    score = minimax(board, depth+1, True)
                    board[i][j] = ' '
                    best_score = min(score, best_score)
        return best_score

# Function to find the best move for the AI
def find_best_move(board):
    best_score = float('-inf')
    best_move = None
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                board[i][j] = 'O'
                score = minimax(board, 0, False)
                board[i][j] = ' '
                if score > best_score:
                    best_score = score
                    best_move = (i, j)
    return best_move

# Main function to play the game
def play_game():
    board = [[' ', ' ', ' '], [' ', ' ', ' '], [' ', ' ', ' ']]
    print_board(board)
    while not game_over(board):
        # Player move
        print("Your turn (X)")
        while True:
            row = int(input("Enter a row number (0-2): "))
            col = int(input("Enter a column number (0-2): "))
            if board[row][col] == ' ':
                board[row][col] = 'X'
                break
            else:
                print("That spot is already taken!")
        print_board(board)
        
        # AI move
        if not game_over(board):
            print("AI's turn (O)")
            row, col = find_best_move(board)
            board[row][col] = 'O'
            print_board(board)
        
    # Print the winner
    if check_winner(board, 'X'):
        print("You win!")
    elif check_winner(board, 'O'):
        print("AI wins!")
    else:
        print("It's a tie!")
```
