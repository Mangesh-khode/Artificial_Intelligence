```python
  graph = {'0': {'2': 0.0021, '1': 0.1068, '3': 0.0759,'4': 0, '15': 0.0108},
              '1': {'0': 0.1023, '2': 0.0682, '3': 0.0162},
              '2': {'0': 0.0080, '1': 0.0584, '13': 0.1261, '15': 0.0076, '23': 0.0575},
              '3': {'0': 0.0696, '1': 0.0875, '4': 0.0321, '5': 0.0713, '6': 0.0472, '7': 0.0055},
              '4': {'4': 0.0018, '3': 0.0755, '7': 0.1997, '11': 0.0389},
              '5': {'3': 0.0517, '7': 0.0816, '8': 0.2379},
              '6': {'3': 0.0189, '20': 0.1448},
              '7': {'3': 0.0552, '4': 0.5574, '5':0.1338, '16': 0.0813, '21': 0.2049},
              '8': {'5': 0.1607, '9': 0.0494, '17': 0.0520},
              '9': {'8': 0.0764, '10': 0.1888, '12': 0.0141},
              '10': {'9': 0.4145, '13': 0.0886, '17': 0.2057},
              '11': {'4': 0.0052, '14': 0.0340, '15': 0.0160, '18': 0.0951},
              '12': {'9': 0.0133, '14':0.0352, '17': 0.0964, '19': 0.1484, '26': 0.3061},
              '13': {'2': 0.0600, '10': 0.1121, '23': 0.0957},
              '14': {'11': 0.0177, '12': 0.0900},
              '15': {'0': 0.0053, '2': 0.0130, '11': 0.0299},
              '16': {'7': 0.0910, '18': 0.0148},
              '17': {'8': 0.0508, '10': 0.0965, '12': 0.1103},
              '18': {'11': 0.1452, '16': 0.0428, '19': 0.0066, '21': 1, '25': 0.4324},
              '19': {'12': 0.0717, '18': 0.0022,'20': 0.1258, '22': 0.2731, '24': 0.4461},
              '20': {'6': 0.1328, '19': 0.0774, '21': 0.0449, '24': 0.0561, '26': 0.5055},
              '21': {'7': 0.3664, '18': 0.1833, '20': 0.0526},
              '22': {'19': 0.5428},
              '23': {'2': 0.0858, '13': 0.1497, '26': 0.2210},
              '24': {'19': 0.2451, '20': 0.1177},
              '25': {'18': 0.0963},
              '26': {'12': 0.1358, '20': 0.2177, '23': 0.1737}}
class Node:
    def __init__(self, state, parent=None, cost=0.0):
        self.state = state
        self.parent = parent
        self.cost = cost

def beam_search(graph, start_state, goal_state, beam_width=4):

    frontier = [Node(start_state)]
    explored = set()

    while frontier:
        new_frontier = []

        for node in frontier:
            if node.state == goal_state:
                path = []
                while node:
                    path.append(node.state)
                    node = node.parent
                path.reverse()
                return path
            explored.add(node.state)
            for next_state, cost in graph[node.state].items():
                if next_state in explored:
                    continue

                child = Node(next_state, node, cost + node.cost)
                new_frontier.append(child)

            new_frontier.sort(key=lambda n: n.cost)
            new_frontier = new_frontier[:beam_width]

        if not new_frontier:
            return None

        frontier = new_frontier

goal_state = "20"
path = beam_search(graph, "12", goal_state)

if path:
    print("Path found:", path)
else:
    print("Path not found")

```
