```python
adj_list={
    0:[1,2,3,4,15],
    1:[0,1,2,3],
    2:[0,15,13,23,1],
    3:[0,1,4,5,6,7],
    4:[0,3,4,11],
    5:[3,7,8],
    6:[3,20],
    7:[3,4,5,16,21],
    8:[9,5,17],
    9:[8,12,10],
    10:[13,9,17],
    11:[4,14,15,18],
    12:[9,14,19,17,26],
    13:[2,23,10],
    14:[11,12],
    15:[0,2,11],
    16:[7,18],
    17:[10,8,12],
    18:[11,16,19,25,21],
    19:[12,18,20,24,22],
    20:[6,21,19,24,26],
    21:[7,18,20],
    22:[19],
    23:[2,13,26],
    24:[19,20],
    25:[18],
    26:[12,20,23]
    
}

from queue import Queue
visited={}
level={}
parent={}
bfs_trav=[]
queue=Queue()

for node in adj_list.keys():
    visited[node]=False
    parent[node]=None
    level[node]=-1
    
s=0
visited[s]=True
level[s]=0
queue.put(s)

while not queue.empty():
    u=queue.get()
    bfs_trav.append(u)
    
    for v in adj_list[u]:
        if not visited[v]:
            visited[v]=True
            parent[v]=u
            level[v]=level[u]+1
            queue.put(v)
            
print(bfs_trav)

v=5
path=[]

while v is not None:
    path.append(v)
    v=parent[v]
path.reverse()
print(path)
    
```
