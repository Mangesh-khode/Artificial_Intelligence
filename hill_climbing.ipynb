```python
def solve_n_queens(initial_state):
    def is_safe(board, row, col):
        if any(board[i] == col for i in range(row)):
            return False
        if any(abs(board[i] - col) == abs(i - row) for i in range(row)):
            return False
        return True

    def solve_util(board, row):
        if row == 8:
            return board[:]
        
        for col in range(8):
            if is_safe(board, row, col):
                board[row] = col
                if solve_util(board, row + 1) is not None:
                    return board[:]
                board[row] = -1  

        return None

  
    board = initial_state[:]
    
    solution = solve_util(board, 0)
    
    return solution

def display_board(queens_positions):
    for row in range(8):
        line = ""
        for col in range(8):
            if queens_positions[col] == row:
                line += "Q "
            else:
                line += ". "
        print(line.strip())

initial_state = [0, 0, 0, 0, 0, 0, 0, 0]
solution = solve_n_queens(initial_state)
print("Queens Positions:", solution)
print("\nFinal Chessboard:")
display_board(solution)
```
