```python
class DARTExpertSystem:
    def __init__(self):
        self.questions = [
            "Does the plant have yellow leaves?",
            "Is the plant wilting?",
            "Are there spots on the leaves?",
            "Is there mold or mildew on the plant?",
            "Are the edges of the leaves turning brown?"
        ]
        self.diseases = {
            "drought_stress": ["yellow leaves", "wilting"],
            "fungal_infection": ["spots on the leaves", "mold or mildew"],
            "nutrient_deficiency": ["yellow leaves", "spots on the leaves", "brown leaf edges"],
            "bacterial_infection": ["wilting", "spots on the leaves", "brown leaf edges"],
        }
        self.symptoms = []

    def ask_questions(self):
        for question in self.questions:
            response = input(question + " (yes/no): ").strip().lower()
            if response == "yes":
                symptom = question.split("have ", 1)[-1].rstrip("?")  # Changed to [-1] to get the last item
                self.symptoms.append(symptom)

    def diagnose(self):
        possible_diseases = {}
        for disease, disease_symptoms in self.diseases.items():
            match_count = len(set(disease_symptoms) & set(self.symptoms))
            if match_count > 0:
                possible_diseases[disease] = match_count

        if possible_diseases:
            sorted_diseases = sorted(possible_diseases.items(), key=lambda x: x[1], reverse=True)
            for disease, count in sorted_diseases:
                print(f"Possible disease: {disease} (matched {count} symptoms)")
        else:
            print("No matching diseases found.")

def main():
    system = DARTExpertSystem()
    print("**********Welcome to the Plant Disease DART Expert System!************")
    system.ask_questions()
    system.diagnose()
if __name__ == "__main__":
    main()
```
